```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
#Check the missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
```

```python
dataset
```

```python
dataset["Region"] = dataset["Region"].astype(int)
```

```python
dataset
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset = dataset.dropna().reset_index(drop=True)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
df = dataset.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##Fix spaces in the column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
#Change the require column to integer data type
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
#Change the other columns to float datatype.
objects = [features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
  if i != 'Classes':
    df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#let's save the clean dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

EXPLORATORY ANALYSIS

```python
df_copy = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
#Encode Classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 'not fire', 'fire')
```

```python
df_copy.head()
```

```python
#Categories in Classes
df_copy['Classes'].value_counts()
```

```python
import seaborn as sns
sns.set_style('darkgrid')  # or any other seaborn style like 'whitegrid', 'ticks' etc.
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
percentage = df_copy.Classes.value_counts(normalize=True) *100
```

```python
#Plotting piechart
plt.figure(figsize=(10,10))
plt.pie(percentage, labels=['Not Fire', 'Fire'], autopct='%1.2f%%')
plt.title('Percentage of Classes')
plt.show()
```

```python
#Correlation
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr(), annot=True)
```

```python
#Boxplots
sns.boxplot(df_copy['FWI'], color='red')
plt.show()
```

```python
##Monthly Fire analysis
dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=dftemp)
plt.xlabel('Months', weight='bold')
plt.ylabel('Number of fires', weight='bold')
plt.title('Monthly Fire Analysis for Bejaia region')
plt.show()
```

```python
##Monthly Fire analysis
dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=dftemp)
plt.xlabel('Months', weight='bold')
plt.ylabel('Number of fires', weight='bold')
plt.title('Monthly Fire Analysis for Bejaia region')
plt.show()
```
